# Limpieza de datos con Python

## Importar Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')

print("✓ Archivo cargado exitosamente")
print(f"✓ DataFrame creado con {len(df)} filas y {len(df.columns)} columnas")

# Mostrar las primeras 10 filas del DataFrame para confirmar que los datos se han cargado correctamente
print("\nPrimeras 10 filas del DataFrame:")
df.head(10)
```

# Obtener información sobre el data frame

```python
print("Información general del DataFrame (info()):")
print("Incluye: número de entradas, nombres de las columnas, tipos de datos y memoria utilizada")
print("-" * 60)
df.info()
```

# **Eliminar Duplicados**

```python
#1. Eliminar duplicados
df = df.drop_duplicates()
print(f"✓ {len(df)} filas sin duplicados")


# 2. Identificar columnas categóricas
columnas_categoricas = df.select_dtypes(include=['object']).columns
print(f"\n{len(columnas_categoricas)} columnas categóricas identificadas:")
print(list(columnas_categoricas))


# 3. Mostrar valores únicos antes de la limpieza
print("\n--- VALORES ÚNICOS ANTES DE LA LIMPIEZA ---")
for col in columnas_categoricas:
    print(f"\n{col}: {df[col].unique()}")
```

##Solucionar inconsistencias

#  Analizaremos los valores inapropiados de la siguiente manera.






```python
# 4. Eliminar espacios en blanco
df[columnas_categoricas] = df[columnas_categoricas].apply(lambda x: x.str.strip())
print("\nEspacios en blanco eliminados")

# 5. Escribir todo en minúsculas
df[columnas_categoricas] = df[columnas_categoricas].apply(lambda x: x.str.lower())
print("Todo convertido a minúsculas")

# 6. Reemplazar valores vacíos con NaN
df[columnas_categoricas] = df[columnas_categoricas].replace('', np.nan)
print("Valores vacíos reemplazados con NaN")

# 7. Mostrar valores únicos después de la limpieza
print("\n--- VALORES ÚNICOS DESPUÉS DE LA LIMPIEZA ---")
for col in columnas_categoricas:
    print(f"\n{col}: {df[col].unique()}")

# 8. Información sobre valores nulos
print("\n--- INFORMACIÓN DE VALORES NULOS ---")
for col in columnas_categoricas:
    nulos = df[col].isnull().sum()
    if nulos > 0:
        print(f"{col}: {nulos} valores nulos")
```

```python
# 9. ANÁLISIS DE VALORES FALTANTES
print("\n" + "="*50)
print("ANÁLISIS DE VALORES FALTANTES")
print("="*50)

# Identificar valores faltantes en todo el dataset
print("\n--- VALORES FALTANTES EN TODO EL DATASET ---")
valores_faltantes = df.isnull().sum()
columnas_con_faltantes = valores_faltantes[valores_faltantes > 0]
print(f"Columnas con valores faltantes: {len(columnas_con_faltantes)}")
for col, count in columnas_con_faltantes.items():
    porcentaje = (count / len(df)) * 100
    print(f"{col}: {count} valores faltantes ({porcentaje:.2f}%)")

# 10. Identificar columnas numéricas
columnas_numericas = df.select_dtypes(include=['int64', 'float64']).columns
print(f"\n{len(columnas_numericas)} columnas numéricas identificadas:")
print(list(columnas_numericas))

# 11. Análisis básico de columnas numéricas
print("\n--- ANÁLISIS DE COLUMNAS NUMÉRICAS ---")
for col in columnas_numericas:
    print(f"\nColumna: {col}")
    print(f"  Tipo de dato: {df[col].dtype}")
    print(f"  Valores únicos: {df[col].nunique()}")
    print(f"  Rango: {df[col].min()} a {df[col].max()}")
    print(f"  Media: {df[col].mean():.2f}")
    print(f"  Mediana: {df[col].median():.2f}")

    # Verificar valores apropiados según el tipo de columna
    if col == 'SeniorCitizen':
        # SeniorCitizen debe ser 0 o 1
        valores_invalidos = df[~df[col].isin([0, 1])]
        if len(valores_invalidos) > 0:
            print(f"  Valores inapropiados encontrados: {valores_invalidos[col].unique()}")
        else:
            print(f"  Valores apropiados (0 o 1)")

    elif col == 'tenure':
        # Tenure debe ser >= 0
        valores_invalidos = df[df[col] < 0]
        if len(valores_invalidos) > 0:
            print(f"  Valores inapropiados encontrados: {valores_invalidos[col].unique()}")
        else:
            print(f"  Valores apropiados (>= 0)")

    elif col == 'MonthlyCharges':
        # MonthlyCharges debe ser >= 0
        valores_invalidos = df[df[col] < 0]
        if len(valores_invalidos) > 0:
            print(f"  Valores inapropiados encontrados: {valores_invalidos[col].unique()}")
        else:
            print(f"  Valores apropiados (>= 0)")

# 12. Análisis de TotalCharges
print(f"\n--- ANÁLISIS DE TotalCharges ---")
print(f"Tipo actual: {df['TotalCharges'].dtype}")
print(f"Valores únicos: {df['TotalCharges'].nunique()}")

# Convertir TotalCharges a numérico
df['TotalCharges_numeric'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
valores_no_numericos = df[df['TotalCharges_numeric'].isnull()]['TotalCharges'].unique()
if len(valores_no_numericos) > 0:
    print(f"Valores no numéricos en TotalCharges: {valores_no_numericos}")
else:
    print(f"Todos los valores son numéricos")

print(f"\nTotalCharges después de conversión:")
print(f"  Rango: {df['TotalCharges_numeric'].min():.2f} a {df['TotalCharges_numeric'].max():.2f}")
print(f"  Media: {df['TotalCharges_numeric'].mean():.2f}")
print(f"  Mediana: {df['TotalCharges_numeric'].median():.2f}")
```

Al tener 0,16 % de valores faltantes decidimos eliminarlos

```python

# 13. Eliminar valores faltantes
df = df.dropna()
print(f"\n{len(df)} filas después de eliminar valores faltantes")

# 14. Convertir TotalCharges definitivamente
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')


# 15. Resumen final
print(f"\n" + "="*50)
print("RESUMEN FINAL")
print("="*50)
print(f"Filas finales: {len(df)}")
print(f"Columnas totales: {len(df.columns)}")
print(f"Valores faltantes: {df.isnull().sum().sum()}")
print(f"Tipos de datos:")
print(df.dtypes.value_counts())

# 16. Mostrar las primeras filas del dataset limpio
print(f"\n--- PRIMERAS 5 FILAS DEL DATASET LIMPIO ---")
print(df.head())
```

```python
print("\n2. GRÁFICO DE BARRAS - Distribución de Churn")

# Contar valores de Churn
churn_counts = df['Churn'].value_counts()
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(8, 6))
bars = plt.bar(churn_counts.index, churn_counts.values, color=colors, alpha=0.7, edgecolor='black')

# Agregar etiquetas con valores
for bar, count in zip(bars, churn_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
             str(count), ha='center', va='bottom', fontweight='bold')

plt.title('Distribución de Abandono de Clientes (Churn)', fontsize=14, fontweight='bold')
plt.xlabel('Churn', fontsize=12)
plt.ylabel('Número de Clientes', fontsize=12)
plt.xticks([0, 1], ['No Abandonó', 'Abandonó'])

# Agregar porcentajes
total = len(df)
for i, count in enumerate(churn_counts.values):
    porcentaje = (count / total) * 100
    plt.text(i, count/2, f'{porcentaje:.1f}%', ha='center', va='center',
             fontweight='bold', fontsize=12)

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('grafico_barras_churn.png', dpi=300, bbox_inches='tight')
plt.show()
```

1. La mayoria de los clientes no abandonan el servicio 73,5%.
2. Aproximadamente 26,5% de los clientes abandona
3. Se debe identificar factores que causan el abandono



```python
print("\n" + "="*60)
print("VISUALIZACIONES MULTIVARIADAS")
print("="*60)
print("\n3. GRÁFICO DE BARRAS AGRUPADAS - Churn por Contract")

# Crear tabla de contingencia
contingency_table = pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100

plt.figure(figsize=(10, 6))

# Configurar el gráfico de barras
x = np.arange(len(contingency_table.index))
width = 0.35

bars1 = plt.bar(x - width/2, contingency_table['no'], width,
                label='No Abandonó', color='lightgreen', alpha=0.7, edgecolor='black')
bars2 = plt.bar(x + width/2, contingency_table['yes'], width,
                label='Abandonó', color='lightcoral', alpha=0.7, edgecolor='black')

plt.title('Porcentaje de Abandono por Tipo de Contrato', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Contrato', fontsize=12)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.xticks(x, contingency_table.index)
plt.legend()
plt.grid(True, alpha=0.3)

# Agregar etiquetas con valores
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('barras_agrupadas_churn_contract.png', dpi=300, bbox_inches='tight')
plt.show()
```

1. Los contratos mes a mes tienen la mayor tasa de abandono 42,7%
2. Los contratos de dos años tienen la menor tasa de abandono 2,8%
3. Los contratos de un año tienen una tasa intermedia 11,1%
4. Los contratos más largos ayudan a retener clientes

```python
print("\n4. GRÁFICO DE BARRAS - Tasa de Churn por Tipo de Internet")

# Crear tabla de contingencia para InternetService
internet_churn = pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100

plt.figure(figsize=(10, 6))

# Crear gráfico de barras
x = np.arange(len(internet_churn.index))
width = 0.35

bars1 = plt.bar(x - width/2, internet_churn['no'], width,
                label='No Abandonó', color='lightgreen', alpha=0.7, edgecolor='black')
bars2 = plt.bar(x + width/2, internet_churn['yes'], width,
                label='Abandonó', color='lightcoral', alpha=0.7, edgecolor='black')

plt.title('Tasa de Abandono por Tipo de Servicio de Internet', fontsize=14, fontweight='bold')
plt.xlabel('Tipo de Servicio de Internet', fontsize=12)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.xticks(x, internet_churn.index)
plt.legend()
plt.grid(True, alpha=0.3)

# Agregar etiquetas con valores
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 1,
                f'{height:.1f}%', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('barras_churn_internet.png', dpi=300, bbox_inches='tight')
plt.show()
```

1. Los clientes sin internet tienen la menor tasa de abandono 7,6%
2. Los clientes con fibra optica tienen la mayor tasa de abandono 41,9%
3. Los clientes con DSL tienen una tasa intermedio 19,0%
4. el tipo de servicio de internet influye significativamente en el abandono
